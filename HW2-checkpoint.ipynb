<h3> Task 1: Load Intrusion.csv </h3>

```python
import numpy as np
from matplotlib import pyplot as plt
import math
import csv
import statistics as stat
```

```python
#arrays for intrusion data
attribute1 = []
attribute2 = []
attribute3 = []
attributes = []
intrusion = []
intrusionColor = []
first = True
#load intrusion.csv
with open('intrusion.csv', mode = 'r')as file:
    
    csvFile = csv.reader(file)
    for lines in csvFile:
        if(not first):
            attribute1.append(float(lines[0]))
            attribute2.append(float(lines[1]))
            attribute3.append(float(lines[2]))
            attributes.append([lines[0],lines[1],lines[2]])
            intrusion.append(float(lines[3]))
            if(lines[3] == '0'):
                intrusionColor.append("blue")
            else:
                intrusionColor.append("red")
        else:
            first = False
```

<h3> Task 2: Create Scatter Plots </h3>

```python
plt.scatter(attribute1,attribute2,c = intrusionColor)
plt.xlabel("Attribute 1")
plt.ylabel("Attribute 2")
plt.show()
```

```python
plt.scatter(attribute2,attribute3,c = intrusionColor)
plt.xlabel("Attribute 2")
plt.ylabel("Attribute 3")
plt.show()
```

<h3>Task 3: Split Data into Train and Test Sets</h3>

```python
from sklearn.svm import SVC
from sklearn.decomposition import PCA as RandomizedPCA
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report

pca = RandomizedPCA(n_components=150, whiten=True, random_state=42)
svcP = SVC(kernel='poly', class_weight='balanced')
svcL = SVC(kernel='linear', class_weight='balanced')
svcR = SVC(kernel='rbf', class_weight='balanced')
model1 = make_pipeline(pca, svcP)
model2 = make_pipeline(pca, svcL)
model3 = make_pipeline(pca, svcR)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(attributes, intrusion,
                                                random_state=42)
```

<h3>Task 4: Train SVCs using Poly, Linear, RBF</h3>

```python
modelP = svcP.fit(Xtrain, ytrain)
modelL = svcL.fit(Xtrain, ytrain)
modelR = svcR.fit(Xtrain, ytrain)
```

<h3>Task 5: Output Confusion Matrices and Classification Reports</h3>

```python
yfit = svcP.predict(Xtest)
from sklearn.metrics import confusion_matrix
confused=confusion_matrix(ytest,yfit)
tn, fp, fn, tp = confused.ravel()
print("TN: ", tn,"FP: ", fp,"FN: ", fn,"TP: ", tp)
print(confused)
print('---------------Classification Report---------------')
print(classification_report(ytest,yfit))
```

```python
yfit = svcL.predict(Xtest)
confused=confusion_matrix(ytest,yfit)
tn, fp, fn, tp = confused.ravel()
print("TN: ", tn,"FP: ", fp,"FN: ", fn,"TP: ", tp)
print(confused)
print('---------------Classification Report---------------')
print(classification_report(ytest,yfit))
```

```python
yfit = svcR.predict(Xtest)
confused=confusion_matrix(ytest,yfit)
tn, fp, fn, tp = confused.ravel()
print("TN: ", tn,"FP: ", fp,"FN: ", fn,"TP: ", tp)
print(confused)
print('---------------Classification Report---------------')
print(classification_report(ytest,yfit))
```
