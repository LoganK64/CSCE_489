<h3> Task 1: Load in Data </h3>

```python
import numpy as np
from matplotlib import pyplot as plt
import math
import csv
import statistics as stat
```

```python
#vars for statistical measures
lengths = []
diameters = []
heights = []
rings = []

#vars for histograms
mLengths = []
fLengths = []
iLengths = []
mDiameters = []
fDiameters = []
iDiameters = []
mHeights = []
fHeights = []
iHeights = []
mWeights = []
fWeights = []
iWeights = []

with open('abalone.data', mode = 'r')as file:
    csvFile = csv.reader(file)
    for lines in csvFile:
        lengths.append(float(lines[1]))
        diameters.append(float(lines[2]))
        heights.append(float(lines[3]))
        rings.append(int(lines[8]))
        if(lines[0] == 'M'):
            mLengths.append(float(lines[1]))
            mDiameters.append(float(lines[2]))
            mHeights.append(float(lines[3]))
            mWeights.append(float(lines[4]))
        elif(lines[0] == 'F'):
            fLengths.append(float(lines[1]))
            fDiameters.append(float(lines[2]))
            fHeights.append(float(lines[3]))
            fWeights.append(float(lines[4]))
        else:
            iLengths.append(float(lines[1]))
            iDiameters.append(float(lines[2]))
            iHeights.append(float(lines[3]))
            iWeights.append(float(lines[4]))
```

<h3> Task 2: Calculate Min, Max, Mean, and Standard Deviation </h3>

<h5> Calculate Min </h5>

```python
print("Min Length: ", min(lengths, key = float))
print("Min Diameter: ", min(diameters, key = float))
print("Min Height: ", min(heights, key = float))
print("Min Rings: ", min(rings, key = int))
```

<h5> Calculate Max </h5>

```python
print("Max Length: ", max(lengths, key = float))
print("Max Diameter: ", max(diameters, key = float))
print("Max Height: ", max(heights, key = float))
print("Max Rings: ", max(rings, key = int))
```

<h5> Calculate Mean </h5>

```python
print("Mean Length:", "{:.3f}".format(stat.mean(lengths)))
print("Mean Diameter: ", "{:.3f}".format(stat.mean(diameters)))
print("Mean Height: ", "{:.3f}".format(stat.mean(heights)))
print("Mean Rings: ", "{:.3f}".format(stat.mean(rings)))
```

<h5> Calculate Standard Deviation </h5>

```python
print("Std Dev Length:", "{:.3f}".format(np.std(lengths)))
print("Std Dev Diameter: ", "{:.3f}".format(np.std(diameters)))
print("Std Dev Height: ", "{:.3f}".format(np.std(heights)))
print("Std Dev Rings: ", "{:.3f}".format(np.std(rings)))
```
